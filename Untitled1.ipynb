```python
import pandas as pd
import numpy as np
import re
#import math
import time
#%matplotlib inline
from matplotlib import pyplot as plt
plt.style.use('ggplot')
import seaborn as sns 
sns.set()
import plotly.plotly as py
import plotly.graph_objs as go
from nltk.corpus import stopwords
from wordcloud import WordCloud
from textblob import TextBlob
import nltk
nltk.download('stopwords')
```

```python
reviews=pd.read_csv('indeed_reviews.csv')
```

```python
reviews['tot_review']=reviews.tot_review/20
```

```python
reviews['review_content']=reviews['review_content'].fillna('')

reviews['review_content'] = reviews['review_content'].apply(lambda x: " ".join(x.lower() for x in x.split()))
# \S+ means anything that is not an empty space
reviews['review_content'] = reviews['review_content'].apply(lambda x: re.sub('http\S*', '', x))
# \s+ means all empty space (\n, \r, \t)
reviews['review_content'] = reviews['review_content'].apply(lambda x: re.sub('\s+', ' ', x))
# We don't want empty string in our text
reviews = reviews.loc[reviews['review_content'] != ""]

reviews['review_content'] = reviews['review_content'].apply(lambda x: re.sub('[^\w\s]','', x))

stop = stopwords.words('english')

stop = stop 
print(stop)
```

```python


def sentiment_func(x):
    sentiment = TextBlob(x['review_content'])
    x['polarity'] = sentiment.polarity
    x['subjectivity'] = sentiment.subjectivity
    return x

reviews= reviews.apply(sentiment_func, axis=1)
```

```python
plt.scatter(reviews.polarity,reviews.tot_review)
```

```python
from sklearn.metrics import r2_score
r2_score(reviews.polarity,reviews.tot_review)
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(reviews[['polarity']],reviews['tot_review'])
         

# Make predictions using the testing set
diabetes_y_pred = regr.predict(reviews[['polarity']])

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(reviews.tot_review, diabetes_y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(reviews.tot_review, diabetes_y_pred))

#ax=sns.lmplot([1,2,3],[4,5,6])

ax=sns.lmplot(x='polarity', y='tot_review', data=reviews,y_jitter=0.1,size=8,aspect=2,
              fit_reg=False,scatter_kws={"s": 20,'alpha':0.2})
ax.set_axis_labels('Polarity', 'Total review')
ax.savefig("correlation.png")
#ax = sns.scatterplot(x="total_bill", y="tip",
#...                      hue="time", style="time", data=tips)

# Plot outputs
#plt.scatter(reviews.polarity,reviews.tot_review,  color='black').
#plt.plot(reviews[['polarity']], diabetes_y_pred, color='blue', linewidth=3)

#plt.xticks(())
#plt.yticks(())
```
